```python
print("\n----- Task 2: Pass / Fail Analyzer -----")

marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
print("\n----- Task 3: Simple Data Cleaner -----")

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned = name.strip().lower()
    cleaned_names.append(cleaned)

print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
print("\nTask 4 - Message Length Analyzer")

messages = ["Hi", "Welcome to the platform", "OK"]

for m in messages:
    print("Message:", m)
    print("Length:", len(m))

    if len(m) > 10:
        print("This message is longer than 10 characters")
    print()
```
